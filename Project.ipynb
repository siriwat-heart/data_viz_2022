##รวมตาราง Covid

```python
data_covid = pd.concat([data_covid1,data_covid2,data_covid3,data_covid4])
data_covid
```

```python
covid = data_covid[['No.','announce_date','sex','age','Unit','nationality','risk','province_of_isolation']]
covid
```

###ดูค่า Missing ของ data_covid

```python
covid.isnull().any() ##ดูค่า Missing ในตาราง data_covid
```

### กำจัดค่า Missing 

```python
set(covid['province_of_isolation']) #ตรวจสอบค่าใน province_of_isolation
```

```python
covid['province_of_isolation'].replace({'กทม':'กรุงเทพมหานคร','ขอนแก่น\xa0':'ขอนแก่น','ภููเก็ต':'ภูเก็ต'}, inplace=True ) #แทนที่ข้อมูลที่ผิดปกติ
```

```python
covid[['sex']] = covid[['sex']].fillna('ไม่ระบุ') ##แทนค่า NaN เป็น ไม่ระบุ
covid[['Unit']] = covid[['Unit']].fillna('ปี') ##แทนค่า NaN ใน Unit เป็น ปี
covid[['nationality']] = covid[['nationality']].fillna('ไม่ระบุ') ##แทนค่า NaN เป็น ไม่ระบุ
covid[['province_of_isolation']] = covid[['province_of_isolation']].fillna('ไม่ระบุ') ##แทนค่า NaN เป็น ไม่ระบุ
covid[['risk']] = covid[['risk']].fillna('ไม่ระบุ') ##แทนค่า NaN เป็น ไม่ระบุ
```

```python
covid.isnull().any() ##ดูค่า Missing ในตาราง data_covid
```
